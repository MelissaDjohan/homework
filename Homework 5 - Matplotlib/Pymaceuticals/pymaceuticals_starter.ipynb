```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinic_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data_df = pd.merge(mouse_data, clinic_data, )

# Display the data table for preview
data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time_gb = data_df.groupby(['Drug', 'Timepoint'])
mean_vol = drug_time_gb["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
mean_df = pd.DataFrame(mean_vol).reset_index()

# Preview DataFrame
mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem = drug_time_gb["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
sem_df = pd.DataFrame(sem).reset_index()

# Preview DataFrame
sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
minor_munging = mean_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
minor_munging
```

```python
# Generate the Plot (with Error Bars):

# Locate
capo_error = sem_df.loc[sem_df["Drug"] == "Capomulin", "Tumor Volume (mm3)"]
infu_error = sem_df.loc[sem_df["Drug"] == "Infubinol", "Tumor Volume (mm3)"]
keta_error = sem_df.loc[sem_df["Drug"] == "Ketapril", "Tumor Volume (mm3)"]
plac_error = sem_df.loc[sem_df["Drug"] == "Placebo", "Tumor Volume (mm3)"]

# Define 'Time'
Time = [0,5,10,15,20,25,30,35,40,45]

# website for markers chart: https://matplotlib.org/3.1.1/api/markers_api.html
plt.errorbar(Time, minor_munging["Capomulin"] , yerr= capo_error, label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.errorbar(Time, minor_munging["Infubinol"] , yerr= infu_error, label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.errorbar(Time, minor_munging["Ketapril"] , yerr= keta_error, label= "Ketapril", marker= "s", color="green", linestyle='--')
plt.errorbar(Time, minor_munging["Placebo"] , yerr= plac_error , label= "Placebo", marker= "D", color="black", linestyle='--')

# Format chart
plt.legend()
plt.title("Tumor Response to Treatment ")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()


# Save the Figure
plt.savefig("../Images/Tumor_Response_Chart.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met = drug_time_gb["Metastatic Sites"].mean()

# Convert to DataFrame
mean_met_df = pd.DataFrame(mean_met).reset_index()

# Preview DataFrame
mean_met_df.head()
```

```python
# Generate the Plot (with Error Bars):

# Locate
capo_met_error = sem_met_df.loc[sem_met_df["Drug"] == "Capomulin", "Metastatic Sites"]
infu_met_error = sem_met_df.loc[sem_met_df["Drug"] == "Infubinol", "Metastatic Sites"]
keta_met_error = sem_met_df.loc[sem_met_df["Drug"] == "Ketapril", "Metastatic Sites"]
plac_met_error = sem_met_df.loc[sem_met_df["Drug"] == "Placebo", "Metastatic Sites"]

# Define 'Time'
Time = [0,5,10,15,20,25,30,35,40,45]

# website for markers chart: https://matplotlib.org/3.1.1/api/markers_api.html
plt.errorbar(Time, minor_met_munging["Capomulin"] , yerr= capo_met_error, label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.errorbar(Time, minor_met_munging["Infubinol"] , yerr= infu_met_error, label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.errorbar(Time, minor_met_munging["Ketapril"] , yerr= keta_met_error, label= "Ketapril", marker= "s", color="green", linestyle='--')
plt.errorbar(Time, minor_met_munging["Placebo"] , yerr= plac_met_error , label= "Placebo", marker= "D", color="black", linestyle='--')

# Format chart
plt.legend()
plt.title("Metastatic Spread During Treatment ")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.grid()


# Save the Figure
plt.savefig("../Images/Metastatic_Spread_Chart.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = drug_time_gb["Mouse ID"].count()

# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count).reset_index()

# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
minor_mice_munging = mice_count_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Mouse ID')

# Preview the Data Frame
minor_mice_munging.head()
```

```python
# Generate the Plot (Accounting for percentages):

# Define 'Time'
Time = [0,5,10,15,20,25,30,35,40,45]

# Plot
plt.plot(Time, (minor_mice_munging["Capomulin"]/25)*100 , label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.plot(Time, (minor_mice_munging["Infubinol"]/25)*100 , label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.plot(Time, (minor_mice_munging["Ketapril"]/25)*100 , label= "Ketapril", marker= "s", color="green", linestyle='--')
plt.plot(Time, (minor_mice_munging["Placebo"]/25)*100 , label= "Placebo", marker= "D", color="black", linestyle='--')

# Format Chart
plt.legend()
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%) ")
plt.grid()


# Save the Figure
plt.savefig("../Images/Survival_Rate_Chart.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug:

# At timepoint 0 for each drug, the volume size is 45. So set starting_vol to 45.
starting_vol = 45

# To locate each final_vol
final_vol = minor_munging.loc[45,:]

# Percent Change is calculated by (final_vol - starting_vol)/starting_vol
perc_chng = 100*(final_vol - starting_vol)/starting_vol


# Display the data to confirm
perc_chng
```

```python
# Store all Relevant Percent Changes into a Tuple:
perc_chng_tup = (perc_chng["Capomulin"],
                perc_chng["Infubinol"],
                perc_chng["Ketapril"],
                perc_chng["Placebo"])

# Splice the data between passing and failing drugs:
fig, summary = plt.subplots()
ind = np.arange(len(perc_chng_tup))  
width = 1
rectsPass = summary.bar(ind[0], perc_chng_tup[0], width, color='green')
rectsFail = summary.bar(ind[1:], perc_chng_tup[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
summary.set_ylabel('% Tumor Volume Change')
summary.set_title('Tumor Change Over 45 Day Treatment')
summary.set_xticks(ind + 0.5)
summary.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
summary.set_autoscaley_on(False)
summary.set_ylim([-30,70])
summary.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        summary.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")

def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        summary.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")

# Call functions to implement the function calls
autolabelPass(rectsPass)
autolabelFail(rectsFail)

# Save the Figure
plt.savefig("../Images/Summary_Chart.png")

# Show the Figure
fig.show()
```
